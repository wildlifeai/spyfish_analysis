```python
tapuae_df = pd.read_excel('MPAMAR Data BUV Tapuae MR 2011 - DOC-1243658.xlsx')
```

```python
cols = tapuae_df.columns
cols = [x.strip() for x in cols]
cols
```

```python
import math
curr_site = None
curr_time = None
l = []
lenl = -1
for row in tapuae_df.iterrows():
    if type(row[1][6]) == str: 
        if lenl == len(l):
            l.append([curr_site, None,  None, None])
        curr_site = row[1][6]
        lenl = len(l) 
    # get current time 
    # print(row[1][11])
    # print(type(row[1][11]))
    if type(row[1][11]) == float:
        curr_time = row[1][11]
        # print(curr_time, row[0])
    for col in range(12,78):
        if type(row[1][col]) != str and not math.isnan(row[1][col]) and not math.isnan(row[1][11]):
            l.append([curr_site, cols[col], curr_time, row[1][col]])

df  = pd.DataFrame(l, columns= ["SiteID", "fishName", "timeInterval","MaxInterval"])
df.drop(index=df.index[0], axis=0, inplace=True)
# df.drop(columns=["index"], inplace=True)
df
```

```python
curr_site = None
curr_time = None
l = []
lenl = -1
prev_site = ""
for row in okura_df.iterrows():
    curr_time = row[1]["Video Time"]
    
    curr_site = row[1]["Site name/num"]
    # check here when it's null deployment
    if curr_site != prev_site:
        if lenl == len(l):
             l.append([curr_site, curr_time,  None, None]) #??? add curr_time or not?
        prev_site = curr_site
        lenl = len(l)

    for fish in [ "snapper #",	"silver trevally #", "goatfish #", "spotty #"]:
        if row[1][fish] != 0:
            l.append([curr_site , fish[:-2], curr_time, row[1][fish]])
    

df  = pd.DataFrame(l, columns= ["SiteID", "fishName", "timeInterval","MaxInterval"])
df.drop(index=df.index[0], axis=0, inplace=True)
df
```
